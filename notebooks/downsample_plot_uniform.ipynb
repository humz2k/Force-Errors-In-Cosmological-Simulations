```python
import sim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from pretty_plots import plot_pretty
```

```python
plot_pretty()
```

```python
n = 1000000
r = 1

ray = sim.utils.ray(np.array([1,0,0]),r*2,25)
analytics = sim.halos.Analytic.Uniform(ray,r=r)
rs = sim.utils.points2radius(ray)

means = []
maxs = []
variances = []
ns = []
downsamples = 2**np.arange(0,16)

for i in range(5):
    df = sim.halos.Sample.Uniform(n,r=r)
    temp_means = []
    temp_maxs = []
    temp_variances = []
    for downsample in tqdm(downsamples):
        downsampled_df = sim.halos.downsample(df,downsample)
        ns.append(len(downsampled_df.index))
        out,time = sim.static_solver.evaluate(ray,downsampled_df,precision="f8")
        diff = out-analytics
        temp_variances.append(np.var(diff))
        rdiff = np.abs(diff)/np.abs(analytics)
        temp_means.append(np.mean(rdiff))
        temp_maxs.append(np.max(rdiff))
    means.append(temp_means)
    maxs.append(temp_maxs)
    variances.append(temp_variances)

print(max(ns),"to",min(ns))
```

```python
for mean in means:
    plt.scatter(downsamples,mean)
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r"$\frac{|\phi_{sum} - \phi_{analytic}|}{|\phi_{analytic}|}$")
plt.xlabel(r"Downsample Amount")
plt.title("Uniform Profile\nMean Error vs Downsample\nN = 1e6, Double Precision",size=10)
plt.tight_layout()
plt.savefig("plots/downsample/UNIFORM_mean_error_f8.jpg")
plt.show()
```

```python
for this_max in maxs:
    plt.scatter(downsamples,this_max)
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r"$\frac{|\phi_{sum} - \phi_{analytic}|}{|\phi_{analytic}|}$")
plt.xlabel(r"Downsample Amount")
plt.title("Uniform Profile\nMax Error vs Downsample\nN = 1e6, Double Precision",size=10)
plt.tight_layout()
plt.savefig("plots/downsample/UNIFORM_max_error_f8.jpg")
plt.show()
```

```python
for variance in variances:
    plt.scatter(downsamples,variance)
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r"$Var({\phi_{sum} - \phi_{analytic}})$")
plt.xlabel(r"Downsample Amount")
plt.title("Uniform Profile\nVariance vs Downsample\nN = 1e6, Double Precision",size=10)
plt.tight_layout()
plt.savefig("plots/downsample/UNIFORM_var_f8.jpg")
plt.show()
```
