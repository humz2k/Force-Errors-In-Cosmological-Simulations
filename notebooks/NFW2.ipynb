```python
import sim
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import pretty_plots
```

```python
ray = sim.utils.ray(np.array([1,0,0]),2,25)
analytic = sim.halos.Analytic.NFW(ray,Rs=1)
```

```python
noises = []
noises_f2 = []
noises_f4 = []
ns = 2**np.arange(5,23)
eps = 1/10
xs = []
for n in tqdm(ns):
    for i in range(3):
        df = sim.halos.Sample.NFW(n,Rs=1)
        outf8,timef8 = sim.static_solver.evaluate(ray,df,precision="f8",eps=eps)
        noises.append((outf8 - analytic))

        #outf4,timef4 = sim.static_solver.evaluate(ray,df,precision="f4-smcuda")
        #noises_f4.append((outf4 - analytic))

        outf2,timef2 = sim.static_solver.evaluate(ray,df,precision="f2-smcuda",eps=eps)
        noises_f2.append((outf2 - analytic))

        xs.append(n)
```

```python
noises_f8 = np.array(noises)
noises_f4 = np.array(noises_f4)
noises_f2 = np.array(noises_f2)
print(noises_f8.shape)
plt.scatter(xs,np.mean((noises_f8/analytic),axis=1))
plt.xscale('log')
plt.show()
```

```python
rels = np.abs((noises_f8-noises_f2))
for noise,n in zip(rels,xs):
    plt.scatter(sim.utils.points2radius(ray),noise,label=str(n))
plt.yscale('log')
plt.show()
```

```python
means = []
for noise in rels:
    means.append(np.mean(np.abs(noise)))
plt.xscale('log')
plt.yscale('log')
plt.scatter(xs,means)
plt.show()
```

```python
noise_mean = []
for noise in noises_f8:
    noise_mean.append(np.mean(np.abs(noise)))
plt.xscale('log')
plt.yscale('log')
plt.scatter(xs,noise_mean, label="(float8 - analytic)")
plt.scatter(xs,means, label="(float8 - analytic) - (float2 - analytic)")
plt.ylabel("Mean Error")
plt.xlabel("N")
plt.legend()
plt.show()
```
